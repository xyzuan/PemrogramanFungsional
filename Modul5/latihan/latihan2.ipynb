```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8, 10])
ypoints = np.array([3, 10, 5])

plt.figure(figsize=(5, 5))
plt.plot(xpoints, ypoints, label="Garis", marker="*", color="black", ls="dotted", linewidth="1")

for i, j in zip(xpoints, ypoints):
   plt.text(i, j+0.5, '({}, {})'.format(i, j))

plt.xlim([0, 16])
plt.ylim([0, 16])
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.legend()
plt.show()
```
