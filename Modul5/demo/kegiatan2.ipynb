```python
import matplotlib.pyplot as plt
import numpy as np

data_transaksi = [
    ("Produk A", 50, 10),
    ("Produk B", 30, 25),
    ("Produk C", 20, 30),
    ("Produk D", 60, 8),
    ("Produk E", 40, 15),
    ("Produk F", 70, 5),
]

harga_produk = list(map(lambda x: x[1], data_transaksi))
jumlah_terjual = list(map(lambda x: x[2], data_transaksi))

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Hubungan Harga Produk dan Jumlah Produk Terjual')
plt.scatter(harga_produk, jumlah_terjual, color='blue')
plt.xlabel('Harga Produk')
plt.ylabel('Jumlah Produk Terjual')

pendapatan_produk = list(map(lambda x: x[1] * x[2], data_transaksi))

plt.subplot(1, 2, 2)
plt.title('Pendapatan Produk')
bar_width = 0.8
index = np.arange(len(data_transaksi))
bars = plt.bar(index, pendapatan_produk, bar_width, label='Pendapatan')
plt.xlabel('Produk')
plt.ylabel('Pendapatan')
plt.xticks(index, [produk[0] for produk in data_transaksi])
plt.legend()

plt.tight_layout()
plt.show()
```
